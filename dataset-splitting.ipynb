Load Dataset



```python
from datasets import load_dataset

# Load the AI4Privacy dataset
dataset = load_dataset("ai4privacy/pii-masking-200k", split="train")
print(dataset[0])   # Inspect first row
```

Split into Train / Validation / Test

```python
# 10,000 rows for training
train_dataset = dataset.select(range(10000))

# 2,000 rows for validation
val_dataset   = dataset.select(range(10000, 12000))

# 1,000 rows for testing
test_dataset  = dataset.select(range(12000, 13000))
```

Keep Only Required Columns

```python
def format_example(example):
    return {
        "source_text": example["source_text"],
        "target_text": example["target_text"],
        "privacy_mask": example["privacy_mask"]
    }

train_dataset = train_dataset.map(format_example, remove_columns=train_dataset.column_names)
val_dataset   = val_dataset.map(format_example, remove_columns=val_dataset.column_names)
test_dataset  = test_dataset.map(format_example, remove_columns=test_dataset.column_names)
```

Save as JSONL




```python
import json

def save_jsonl(dataset, filename):
    with open(filename, "w", encoding="utf-8") as f:
        for ex in dataset:
            f.write(json.dumps(ex) + "\n")

save_jsonl(train_dataset, "/content/drive/MyDrive/train.jsonl")
save_jsonl(val_dataset, "/content/drive/MyDrive/val.jsonl")
save_jsonl(test_dataset, "/content/drive/MyDrive/test.jsonl")



```
